```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
rain_df = pd.read_csv("Resources/avg_rain_state.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
ticks = x_axis
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(15,3))
plt.bar(x_axis, rain_df["Inches"], color='b', alpha=1, align="center")
plt.xticks(ticks, rain_df["State"], rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(-0.4, len(x_axis))
plt.ylim(0, max(rain_df["Inches"])+5)
```

```python
# Set a Title and labels
plt.title("Average Rain by State")
plt.xlabel("State")
plt.ylabel("Average Rainfall in Inches")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("Images/avg_state_rain.png")
plt.show()
```
